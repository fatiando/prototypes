# Some classes for 2D gravity inversion

```python
%matplotlib inline
from __future__ import division
from IPython.html import widgets
from fatiando.gravmag import talwani
from fatiando.inversion.base import Misfit
from fatiando.inversion.regularization import Damping, Smoothness1D, TotalVariation1D, Smoothness2D, Smoothness, fd2d
from fatiando import utils
from fatiando.mesher import Polygon, SquareMesh, Square
from fatiando.vis import mpl
import numpy as np
import scipy.sparse
import multiprocessing
```

## Basin + Polygon

```python
class BasinPoly2D(Misfit):
    def __init__(self, x, z, gz, height, npoints, density):
        super(BasinPoly2D, self).__init__(
            data=gz, nparams=npoints, islinear=False,
            positional=dict(x=x, z=z), 
            model=dict(height=height, 
                       props={'density': density}))
        
    def p2vertices(self, p):
        x, h = self.positional['x'], self.model['height']
        n = self.nparams + 2
        verts = np.empty((n, 2))
        verts[:, 0] = np.linspace(x.min(), x.max(), n)[::-1]
        verts[:, 1] = np.concatenate([[h], p, [h]])
        return verts
    
    def _get_predicted(self, p):
        x, z = self.positional['x'], self.positional['z']
        verts = self.p2vertices(p)
        poly = Polygon(verts, self.model['props'])
        return talwani.gz(x, z, [poly])
        
    def _get_jacobian(self, p):    
        x, z = self.positional['x'], self.positional['z']
        props = self.model['props']
        verts = self.p2vertices(p)
        delta = np.array([0, 10])
        jac = np.empty((self.ndata, self.nparams))
        for i in xrange(self.nparams):
            diff = Polygon([verts[i + 2], verts[i + 1] - delta, 
                            verts[i], verts[i + 1] + delta], props)
            jac[:, i] = talwani.gz(x, z, [diff])/(2 * delta[1])
        return jac
          
    def fit(self):
        super(BasinPoly2D, self).fit()
        self._estimate = Polygon(self.p2vertices(self.p_), 
                                 self.model['props'])
        return self
```

```python
x = np.linspace(0, 100000, 100)
z = -100*np.ones_like(x)
xs = np.linspace(0, 100000, 100)[::-1]
depths = (-1e-15*(xs - 50000)**4 + 8000 
          - 3000*np.exp(-(xs - 70000)**2/(10000**2)))
depths -= depths.min()
density = -500
model = Polygon(np.transpose([xs, depths]), {'density':density})
gz = utils.contaminate(talwani.gz(x, z, [model]), 0.5, seed=10000)
```

```python
mpl.subplot(2, 1, 1)
mpl.plot(x, gz, 'ok')
ax = mpl.subplot(2, 1, 2)
mpl.polygon(model, fill='gray')
ax.invert_yaxis()
```

```python
misfit = BasinPoly2D(x, z, gz, 0, 30, density)
initial = 3000*np.ones(misfit.nparams)
misfit.config('levmarq', initial=initial).fit()
```

```python
mpl.subplot(2, 1, 1)
mpl.plot(x, gz, 'ok')
mpl.plot(x, misfit.predicted(), '-r', linewidth=2)
ax = mpl.subplot(2, 1, 2)
mpl.polygon(model, fill='gray', alpha=0.5)
mpl.polygon(misfit.estimate_, style='o-r')
ax.invert_yaxis()
```

## Linear + Depth Weights

```python
class Blocky2D(Misfit):
    def __init__(self, x, z, data, mesh):
        super(Blocky2D, self).__init__(
            data=data,
            positional=dict(x=x, z=z),
            model=dict(mesh=mesh),
            islinear=True,
            nparams=mesh.size)
    
    def _get_predicted(self, p):
        return  utils.safe_dot(self.jacobian(p), p)
    
    def _get_jacobian(self, p):
        x, z = self.positional['x'], self.positional['z']
        mesh = self.model['mesh']
        jac = np.empty((self.ndata, self.nparams))
        for i in xrange(self.nparams):
            jac[:, i] = talwani.gz(x, z, [mesh[i]], dens=1)
        return jac        
```

```python
area = (0, 50e3, 0, 10e3)
x = np.linspace(area[0], area[1], 200)
z = -100*np.ones_like(x)
model = [Square((20.6e3, 40e3, 3e3, 6.5e3), props=dict(density=500)),
         Square((5.3e3, 10.2e3, 4e3, 8.5e3), props=dict(density=-500))]
gz = utils.contaminate(talwani.gz(x, z, model), 0.1, seed=0)
```

```python
mpl.subplot(2, 1, 1)
mpl.plot(x, gz, '.k')
ax = mpl.subplot(2, 1, 2)
for p in model:
    mpl.polygon(p, fill='gray')
mpl.set_area(area)
ax.invert_yaxis()
```

```python
mesh = SquareMesh(area, (20, 10))
misfit = Blocky2D(x, z, gz, mesh)
solver = (misfit + 1e-10*Smoothness2D(mesh.shape)).fit()
mesh.addprop('density', solver.estimate_)
```

```python
mpl.subplot(2, 1, 1)
mpl.plot(x, gz, '.k')
mpl.plot(x, solver.predicted(), '-r', linewidth=2)
ax = mpl.subplot(2, 1, 2)
for p in model:
    mpl.polygon(p)
mpl.squaremesh(mesh, 'density', cmap=mpl.cm.RdBu_r)
mpl.colorbar(orientation='horizontal', pad=0, aspect=50)
mpl.set_area(area)
ax.set_xticklabels([])
ax.invert_yaxis()
```

```python
def explore_blocky(regularization, mu, nx, nz):
    mesh = SquareMesh(area, (nz, nx))
    misfit = Blocky2D(x, z, gz, mesh)
    if regularization == 'Damping':
        regul = Damping(misfit.nparams)
    elif regularization == 'Smoothness':
        regul = Smoothness2D(mesh.shape)
    solver = (misfit + 10**mu*regul).fit()
    mesh.addprop('density', solver.estimate_)
    mpl.subplot(2, 1, 1)
    mpl.plot(x, gz, '.k')
    mpl.plot(x, solver.predicted(), '-r', linewidth=2)
    ax = mpl.subplot(2, 1, 2)
    for p in model:
        mpl.polygon(p)
    mpl.squaremesh(mesh, 'density', cmap=mpl.cm.RdBu_r)
    mpl.colorbar(orientation='horizontal', pad=0, aspect=50)
    mpl.set_area(area)
    ax.set_xticklabels([])
    ax.invert_yaxis()
```

```python
widgets.interactive(explore_blocky, regularization='Damping Smoothness'.split(), 
                    nx=widgets.IntTextWidget(value=10), nz=widgets.IntTextWidget(value=10),
                    mu=(-20, 0, 0.5))
```

```python
class DepthWeightedSmoothness2D(Smoothness):
    def __init__(self, mesh, power=2, z0=0):
        fdmat = fd2d(mesh.shape)
        super(DepthWeightedSmoothness2D, self).__init__(fdmat)
        self.mesh = mesh
        self.power = power
        self.z0 = z0        
        weights = _depth_weights(mesh, power, z0)
        aux = utils.safe_dot(fdmat, weights)
        self._cache = {}
        self._cache['hessian'] = {'hash':'',
                                  'array':2*utils.safe_dot(aux.T, fdmat)}
        
def _depth_weights(mesh, power, z0):
    w = [(abs(np.mean(c.bounds[-2:])) + z0 + 1e-15)**(-power/2)
         for c in mesh]
    weights = scipy.sparse.diags([w], [0]).tocsr()
    return weights
```

```python
mesh = SquareMesh(area, (20, 20))
misfit = Blocky2D(x, z, gz, mesh)
```

```python
def explore_blocky_dw(mu, power, z0):
    regul = DepthWeightedSmoothness2D(mesh, power, z0)
    solver = (misfit + 10**mu*regul).fit()
    mesh.addprop('density', solver.estimate_)
    mpl.subplot(2, 1, 1)
    mpl.plot(x, gz, '.k')
    mpl.plot(x, solver.predicted(), '-r', linewidth=2)
    ax = mpl.subplot(2, 1, 2)
    for p in model:
        mpl.polygon(p)
    mpl.squaremesh(mesh, 'density', cmap=mpl.cm.RdBu_r)
    mpl.colorbar(orientation='horizontal', pad=0, aspect=50)
    mpl.set_area(area)
    ax.set_xticklabels([])
    ax.invert_yaxis()
    
widgets.interactive(explore_blocky_dw,
                    mu=(-20, 0, 1), power=(0, 6, 1), z0=(0, 10e3, 500))
```
