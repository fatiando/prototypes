```python
%pylab inline
%load_ext cythonmagic
import numpy as np
from fatiando.vis import mpl
```

```python
def hyperbola(x, t0, x0, velocity):
    z0 = t0*velocity
    dx = (x - x0)
    r =  np.sqrt(z0**2 + dx**2)
    time = np.sqrt(t0**2 + dx**2/velocity**2)
    amp = (1./np.sqrt(r*velocity))*(z0/r)
    return time, amp

def time2index(times, dt):
    t = np.floor(times/dt).astype(np.int)
    return t
```

```python
offset = np.linspace(-10000, 10000, 100)
dt = 0.0001
times = np.arange(0, 1, dt)
t, amp = hyperbola(offset, 0.3, 0, 5000)
mpl.figure()
mpl.subplot(1, 2, 1)
mpl.plot(offset, t)
mpl.ylim(times.max(), times.min())
mpl.xlabel('offset')
mpl.ylabel('time')
mpl.subplot(1, 2, 2)
mpl.plot(offset, time2index(t, dt))
mpl.ylim(len(times), 0)
mpl.xlabel('offset')
mpl.ylabel('index')
```

```python
def addhyp(t0, x0, dt, data, amp=1):    
    nt, nx = data.shape
    t, a = hyperbola(offsets, t0, x0, velocity)
    t = time2index(t, dt)
    a /= a.max()
    valid = t < nt
    data[t[valid], valid] += amp*a[valid]
```

```python
shape = (300, 300)
offsets = np.linspace(0, 10000, shape[1])
offset = offsets[1] - offsets[0]
times = np.linspace(0, 1, shape[0])
dt = times[1] - times[0]
velocity = 5000
data = np.zeros(shape)
for t0 in times[(times < 0.32) & (times > 0.3)]:
    addhyp(t0, 2000, dt, data)
for t0 in times[(times < 0.12) & (times > 0.1)]:
    addhyp(t0, 8000, dt, data)
data[100:115, 100:115] = 1
data[40:50, 20:250] = 1

data += np.random.normal(0, 0.1, data.shape)

extent = [offsets[0], offsets[-1], times[-1], times[0]]
plt.figure()
plt.imshow(data, cmap=plt.cm.gray_r, extent=extent, aspect='auto')
```

```python
%%cython
import numpy as np
from libc.math cimport sqrt, floor
cimport cython
from cython.parallel cimport prange

@cython.boundscheck(False)
@cython.wraparound(False)
def kirchhoff(double[:,::1] data, double offset, double dt, double velocity, shape, unsigned int aperture):
    cdef:
        double[:,::1] mig
        unsigned int i, j, nt, nx, t, 
        int k, start, stop
        double t0, x0, dx, amp, z0, r
    mig = np.zeros_like(data)
    nt, nx = shape
    for i in range(nt):
        t0 = i*dt
        z0 = t0*velocity
        for j in range(nx):
            x0 = offset*j
            start = j - aperture
            stop = j + aperture
            if start < 0:
                start = 0
            if stop > nx:
                stop = nx            
            for k in range(start, stop):
                dx = offset*k - x0
                t = int(floor((sqrt(t0**2 + dx**2/velocity**2))/dt))
                if t < nt:                        
                    r =  sqrt(z0**2 + dx**2)
                    amp = 1.
                    if r > 10**(-10):
                        amp *= (1./sqrt(r))*(z0/r)
                    mig[i, j] += data[t, k]*amp
    return mig    
```
