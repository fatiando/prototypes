# Sample implementation for the fatiando.inversion package refactor

```python
%pylab inline
import numpy
import matplotlib.pyplot as plt
import scipy.sparse
from numexpr import evaluate
from fatiando import gravmag, mesher, utils, seismic
from fatiando.vis import mpl
import time
```

```python
def safe_solve(matrix, vector):
    if scipy.sparse.issparse(matrix) or scipy.sparse.issparse(vector):
        estimate, status = scipy.sparse.linalg.cgs(matrix, vector)
        if status >= 0:
            return estimate
        else:
            raise ValueError('CGS exited with input error')
    else:
        return numpy.linalg.solve(matrix, vector)
    
def safe_dot(a, b):
    if scipy.sparse.issparse(a) or scipy.sparse.issparse(b):
        return a*b
    else:
        return numpy.dot(a, b)
    
def safe_diagonal(matrix):
    if scipy.sparse.issparse(matrix):
        return numpy.array(matrix.diagonal())
    else:
        return numpy.diagonal(matrix).copy()
    
def exaust(iterator):
    for things in iterator:
        continue
    return things
```

## Non-linear optimizer

Optimizers receive a list of goal functions, initial estimate and auxiliary variables.
Goal functions know how to compute their value, gradient and Hessian at a given point in parameter space `p`.
Goal functions need to be "moved" to the computation point before getting these values using the `.at(p)` method call.
The `.precompute(p)` method call is when the goal functions can compute intensive tasks (like Jacobian matrix building) in order to not repeat the computation when calculating gradient and Hessian.

```python
def levmarq(goals, initial, maxit=30, maxsteps=10, lamb=10, dlamb=2, tol=10**-5, precondition=True, report=False):
    p = initial
    misfit = sum(g.mu*g.at(p).value() for g in goals)
    for iteration in xrange(maxit):
        [g.precompute() for g in goals]
        if precondition:
            hessian, minus_gradient = None, None
            for g in goals:
                h = g.hessian()
                diag = numpy.abs(safe_diagonal(h))
                diag[diag < 10**-10] = 10**-10
                precond = scipy.sparse.diags(1./diag, 0).tocsr()
                if hessian is None:
                    hessian = g.mu*precond*h
                    minus_gradient = -g.mu*g.gradient()/diag
                else:
                    # Increment breaks for some sparse matrix formats
                    hessian = hessian + g.mu*precond*h
                    minus_gradient = minus_gradient - g.mu*g.gradient()/diag
        else:
            hessian = sum(g.mu*g.hessian() for g in goals)
            minus_gradient = -sum(g.mu*g.gradient() for g in goals)
        stagnation = True
        diagonal = scipy.sparse.diags(safe_diagonal(hessian), 0).tocsr()
        for step in xrange(maxsteps):
            hessian = hessian + lamb*diagonal
            newp = p + safe_solve(hessian, minus_gradient)
            newmisfit = sum(g.mu*g.at(newp).value() for g in goals)
            if newmisfit >= misfit:
                hessian = hessian - lamb*diagonal
                if lamb < 10**15:
                    lamb *= dlamb
            else:
                if lamb > 10**-15:
                    lamb /= float(dlamb)
                stagnation = False
                break
        if stagnation:
            stop = True
        else:
            p = newp
            stop = newmisfit < misfit and abs((newmisfit - misfit)/float(misfit)) < tol
            misfit = newmisfit
        if report:
            yield p, {'iteration':iteration, 'LM steps':step + 1, 'stagnation':stagnation, 'lambda':lamb, 'misfit':misfit}
        else:
            yield p
        if stop:
            break 
```

## Linear solver

The linear overdetermined solver asks for the hessian and the gradient calculated at the null vector:

```python
def overdetermined(goals, precondition=True):
    [g.precompute() for g in goals]
    if precondition:
        hessian, gradient = None, None
        for g in goals:
            h = g.hessian()
            diag = numpy.abs(safe_diagonal(h))
            diag[diag < 10**-10] = 10**-10
            precond = scipy.sparse.diags(1./diag, 0).tocsr()
            if hessian is None:
                hessian = g.mu*safe_dot(precond, h)
                gradient = g.mu*g.gradient_at_null()/diag
            else:
                hessian = hessian + g.mu*safe_dot(precond, h)
                gradient = gradient + g.mu*g.gradient_at_null()/diag
    else:
        hessian = sum(g.mu*g.hessian() for g in goals)
        gradient = sum(g.mu*g.gradient_at_null() for g in goals)
    p = safe_solve(hessian, -gradient)
    return p
```

## Goal functions

An $\ell_2$-norm goal function. This is a generic goal function and needs to be subclassed in order to be used.

```python
class L2Norm(object):
    def __init__(self, data, weights=None):
        super(L2Norm, self).__init__()
        self.data = data
        self.size = len(data)
        self._p = None
        self.mu = 1.
        self._predicted = None
        self._jacobian = None
        self.set_weights(weights)
        
    def set_weights(self, weights):
        if weights is not None:
            self.weights = scipy.sparse.diags(weights, 0)
        else:
            self.weights = None
        
    def at(self, p):
        self._predicted = self.predicted(p)
        self._p = p
        self._jacobian = None
        return self
        
    def precompute(self):
        self._jacobian = self.jacobian(self._p)
        
    def value(self):
        if self.weights is None:            
            return numpy.linalg.norm(self.data - self._predicted)**2/self.size
        else:
            return numpy.sum(self.weights*((self.data - self._predicted)**2))/self.size 
    
    def hessian(self):
        if self.weights is None:
            return 2*safe_dot(self._jacobian.T, self._jacobian)/self.size
        else:
            return 2*safe_dot(self._jacobian.T, self.weights*self._jacobian)/self.size
    
    def gradient(self):
        if self.weights is None:
            grad = -2*safe_dot(self._jacobian.T, self.data - self._predicted)/self.size
        else:
            grad = -2*safe_dot(self._jacobian.T, self.weights*(self.data - self._predicted))/self.size
        # Check if the gradient isn't a one column matrix
        if len(grad.shape) > 1:
            # Need to convert it to a 1d array so that hell won't break loose
            grad = numpy.array(grad.T).ravel()
        return grad
    
    def gradient_at_null(self):
        if self.weights is None:
            grad = -2*safe_dot(self._jacobian.T, self.data)/self.size
        else:
            grad = -2*safe_dot(self._jacobian.T, self.weights*self.data)/self.size
        # Check if the gradient isn't a one column matrix
        if len(grad.shape) > 1:
            # Need to convert it to a 1d array so that hell won't break loose
            grad = numpy.array(grad.T).ravel()
        return grad
    
    def jacobian(self, p):
        raise NotImplementedError("Jacobian matrix not implemented")
    
    def predicted(self, p):
        raise NotImplementedError("Predicted data not implemented") 
```

Regularizing functions are also goal functions:

```python
class Damping(object):
    def __init__(self, mu, nparams):
        super(Damping, self).__init__()
        self.nparams = nparams
        self.mu = mu
        self._p = None
        self._stored_hessian = self._precompute_hessian()
        
    def _precompute_hessian(self):
        return 2*scipy.sparse.identity(self.nparams).tocsr()
        
    def at(self, p):
        self._p = p
        return self
        
    def precompute(self):
        pass
        
    def value(self):
        return numpy.linalg.norm(self._p)**2
    
    def hessian(self):
        return self._stored_hessian
    
    def gradient(self):
        return 2*self._p
    
    def gradient_at_null(self):
        return 0
    
class Smoothness(object):
    def __init__(self, mu, fdmatrix):
        super(Smoothness, self).__init__()
        self.mu = mu
        self._p = None
        self._stored_hessian = self._precompute_hessian(fdmatrix)
        
    def _precompute_hessian(self, fd):
        return 2*fd.T*fd
            
    def at(self, p):
        self._p = p
        return self
        
    def precompute(self):
        pass
        
    def value(self):
        # Need to divide by 2 because the hessian is 2*R.T*R
        return numpy.sum(self._p*safe_dot(self._stored_hessian, self._p))/2.
    
    def hessian(self):
        return self._stored_hessian
    
    def gradient(self):
        # mu is already included in the Hessian
        return safe_dot(self._stored_hessian, self._p)
    
    def gradient_at_null(self):
        return 0
    
class Smoothness1D(Smoothness):
    def __init__(self, mu, nparams):
        super(Smoothness1D, self).__init__(mu, fd1d(nparams))
        self.nparams = nparams
        
class Smoothness2D(Smoothness):
    def __init__(self, mu, shape):
        super(Smoothness2D, self).__init__(mu, fd2d(shape))
        self.shape = shape
        self.nparams = shape[0]*shape[1]
            
class TotalVariation(object):
    def __init__(self, mu, beta, fdmatrix):
        super(TotalVariation, self).__init__()
        self.mu = mu
        self.beta = beta
        self._fdmatrix = fdmatrix
        self._p = None
        
    def at(self, p):
        self._p = p
        return self
    
    def precompute(self):
        pass
        
    def value(self):
        return numpy.linalg.norm(safe_dot(self._fdmatrix, self._p), 1)
    
    def hessian(self):
        derivs = safe_dot(self._fdmatrix, self._p)
        q_matrix = scipy.sparse.diags(self.beta/((derivs**2 + self.beta)**1.5), 0).tocsr()
        return safe_dot(self._fdmatrix.T, q_matrix*self._fdmatrix)
    
    def gradient(self):
        derivs = safe_dot(self._fdmatrix, self._p)
        q = derivs/numpy.sqrt(derivs**2 + self.beta)
        grad = safe_dot(self._fdmatrix.T, q)
        if len(grad.shape) > 1:
            grad = numpy.array(grad.T).ravel()
        return grad
    
    def gradient_at_null(self):
        return 0
    
class TotalVariation1D(TotalVariation):
    def __init__(self, mu, beta, nparams):
        super(TotalVariation1D, self).__init__(mu, beta, fd1d(nparams))
        self.nparams = nparams
    
class TotalVariation2D(TotalVariation):
    def __init__(self, mu, beta, shape):
        super(TotalVariation2D, self).__init__(mu, beta, fd2d(shape))
        self.shape = shape
        self.nparams = shape[0]*shape[1]
        
def fd1d(size):
    i = range(size - 1) + range(size - 1)
    j = range(size - 1) + range(1, size)
    v = [1]*(size - 1) + [-1]*(size - 1)
    return scipy.sparse.coo_matrix((v, (i, j)), (size - 1, size)).tocsr()

def fd2d(shape):
    ny, nx = shape
    nderivs = (nx - 1)*ny + (ny - 1)*nx
    I, J, V = [], [], []
    deriv = 0
    param = 0
    for i in xrange(ny):
        for j in xrange(nx - 1):
            I.extend([deriv, deriv])
            J.extend([param, param + 1])
            V.extend([1, -1])
            deriv += 1
            param += 1
        param += 1
    param = 0
    for i in xrange(ny - 1):
        for j in xrange(nx):
            I.extend([deriv, deriv])
            J.extend([param, param + nx])
            V.extend([1, -1])
            deriv += 1
            param += 1
    return scipy.sparse.coo_matrix((V, (I, J)), (nderivs, nx*ny)).tocsr()
```

## Example use: polynomial regression

This is a problem specific goal function that wraps the data and auxiliary values. It also knows how to compute the Jacobian matrix and predicted data for the case of a second order regression. Since this is a linear problem, the Jacobian can be computed only once.

```python
class PolyData(L2Norm):
    def __init__(self, x, y, weights=None):
        super(PolyData, self).__init__(y, weights)
        self.x = x
        self._stored_jacobian = None
        
    def _calculate_jacobian(self):
        self._stored_jacobian = numpy.transpose([self.x**2, self.x, numpy.ones_like(x)])
                
    def jacobian(self, p):
        if self._stored_jacobian is None:
            self._calculate_jacobian()
        return self._stored_jacobian

    def predicted(self, p):
        return numpy.dot(self._stored_jacobian, p)
```

Testing this on some synthetic data:

```python
x = numpy.linspace(0, 10, 100)
numpy.random.seed(1)
a, b, c = 10, -100, 30000
y = a*x**2 + b*x + c + numpy.random.normal(scale=20, size=len(x))
data = PolyData(x, y)
coefs = overdetermined([data])
plt.figure()
# Use the linear solver
print "LINEAR:"
print 'estimated:', coefs.tolist()
print 'true:', [a, b, c]
plt.plot(x, y, '.k', x, data.predicted(coefs), '-r')
# Run through the non-linear optimizer
print "NON-LINEAR:"
coefs, report = exaust(levmarq([data], numpy.array([0,0,0]), report=True))
print 'estimated:', coefs.tolist()
print 'true:', [a, b, c]
print 'report:', report
plt.plot(x, data.predicted(coefs), '--b')
```

## Example: Gaussian fit

Fit an equation of the form:

$$ f(x) = a\exp(-b(x + c)^{2}) + d\exp(-e(x + f)^{2}) $$

```python
class ExpData(L2Norm):
    def __init__(self, x, y, use_ne=False):
        super(ExpData, self).__init__(y)
        self.x = x
        self._jac = numpy.empty((self.size, 6))
        self.use_ne = use_ne
    
    def jacobian(self, p):
        jac = self._jac
        jac[:,0], jac[:,1], jac[:,2] = self._derivs(p[:3])
        jac[:,3], jac[:,4], jac[:,5] = self._derivs(p[3:])
        return jac
    
    def _derivs(self, coefs):
        a, b, c = coefs
        if self.use_ne:
            x = self.x
            derivs = [
                evaluate('exp(-b*(x + c)**2)'),
                evaluate('(a*exp(-(b + 0.005)*(x + c)**2) - a*exp(-(b - 0.005)*(x + c)**2))/0.01'),
                evaluate('(a*exp(-(b)*(x + c + 0.05)**2) - a*exp(-(b)*(x + c - 0.05)**2))/0.1')]
        else:
            derivs = [
                numpy.exp(-b*(self.x + c)**2),
                # Do a numerical derivative for these two because I'm too lazy to do it by hand
                (a*numpy.exp(-(b + 0.005)*(self.x + c)**2) - a*numpy.exp(-(b - 0.005)*(self.x + c)**2))/0.01,
                (a*numpy.exp(-(b)*(self.x + c + 0.05)**2) - a*numpy.exp(-(b)*(self.x + c - 0.05)**2))/0.1]
        return derivs
    
    def predicted(self, p):
        a, b, c, d, e, f = p
        if self.use_ne:
            x = self.x
            return evaluate('a*exp(-b*(x + c)**2) + d*exp(-e*(x + f)**2)')
        else:
            return a*numpy.exp(-b*(self.x + c)**2) + d*numpy.exp(-e*(self.x + f)**2)
            
```

```python
x = numpy.linspace(0, 10, 1000000)
numpy.random.seed(1)
a, b, c = 100, 0.1, -2
d, e, f = 30, 2, -7
y = a*numpy.exp(-b*(x + c)**2) + d*numpy.exp(-e*(x + f)**2) + numpy.random.normal(scale=5, size=len(x))
data = ExpData(x, y, use_ne=True)
coefs = exaust(levmarq([data], numpy.array([1, 1, 1, 1, 1, -9]), lamb=10**1, maxit=100, tol=10**-8))
print 'estimate:', coefs.tolist()
print 'true:', [a, b, c, d, e, f]
plt.figure()
plt.plot(x, y, '.k', x, data.predicted(coefs), '-')
```

## Example: 2D gravity inversion for the relief of a basin

```python
class BasinPolygon2DGz(L2Norm):
    def __init__(self, x, z, gz, height, npoints, density):
        super(BasinPolygon2DGz, self).__init__(gz)
        self.x, self.z = x, z
        self.height = height
        self.nparams = npoints
        self.shape = (self.size, self.nparams)
        self.props = {'density':density}
        self._left, self._right = [[x.min(), -height]], [[x.max(), -height]]
        self._xs = numpy.linspace(x.min(), x.max(), npoints + 2)[::-1][1:-1]
        
    def toverts(self, p):
        return self._right + numpy.transpose([self._xs, p]).tolist() + self._left
    
    def topolygon(self, p):
        return mesher.Polygon(self.toverts(p), self.props)
    
    def jacobian(self, p):
        verts = self.toverts(p)
        x, z = self.x, self.z
        xs = self._xs
        dz = 10.
        jac = numpy.zeros(self.shape)
        for i in xrange(self.nparams):
            difference = mesher.Polygon([verts[i+2],[xs[i],p[i]-dz],verts[i],[xs[i],p[i]+dz]], self.props)
            jac[:,i] = numpy.nan_to_num(gravmag.talwani.gz(x, z, [difference]))/(2.*dz)
        return jac
        
    def predicted(self, p):
        return gravmag.talwani.gz(self.x, self.z, [self.topolygon(p)])
```

```python
x = numpy.linspace(0, 100000, 100)
z = -100*numpy.ones_like(x)
xs = numpy.linspace(0, 100000, 200)[::-1]
depths = -10**-15*(xs - 50000)**4 + 6000 - 5000*numpy.exp(-(xs - 70000)**2/(10000**2))
depths -= depths.min()
depths[(xs > 30000) & (xs < 40000)] -= 3000
density = -500
model = mesher.Polygon(numpy.transpose([xs, depths]), {'density':density})
gz = utils.contaminate(gravmag.talwani.gz(x, z, [model]), 0.1)
npoints = 100
data = BasinPolygon2DGz(x, z, gz, 0, npoints, density)

mpl.figure(figsize=(10,6))
mpl.subplot(2, 1, 2)
mpl.polygon(model, fill='gray', alpha=0.5)
mpl.ylim(depths.max() + 1000, depths.min())
mpl.xlim(x.min(), x.max())
mpl.m2km()
mpl.subplot(2, 1, 1)
mpl.plot(x*0.001, gz, '.k')
```

Using smoothness constraints:

```python
regularization = [Smoothness1D(1*10**-8, npoints)]
initial = 5000*numpy.ones(npoints)
estimate, report = exaust(levmarq([data] + regularization, initial, report=True))
predicted = data.predicted(estimate)
print 'report:', report

mpl.figure(figsize=(10,6))
mpl.subplot(2, 1, 2)
mpl.title('Smoothness')
mpl.polygon(model, fill='gray', alpha=0.5)
for disturbed in (utils.contaminate(predicted, 0.1) for i in range(5)):
    p = exaust(levmarq([BasinPolygon2DGz(x, z, disturbed, 0, npoints, density)] + regularization, initial))
    mpl.polygon(data.topolygon(p), '-r', linewidth=3, linealpha=0.5)
mpl.polygon(data.topolygon(estimate), '.-b', linewidth=2)
mpl.ylim(depths.max() + 1000, depths.min())
mpl.xlim(x.min(), x.max())
mpl.m2km()
mpl.subplot(2, 1, 1)
mpl.plot(x*0.001, gz, '.k')
mpl.plot(x*0.001, predicted, '-r')
```

```python
regularization = [TotalVariation1D(10**-9, 10**-8, npoints)]
initial = 5000*numpy.ones(npoints)
estimate, report = exaust(levmarq([data] + regularization, initial, report=True))
predicted = data.predicted(estimate)
print 'report:', report

mpl.figure(figsize=(10,6))
mpl.subplot(2, 1, 2)
mpl.title('Total variation')
mpl.polygon(model, fill='gray', alpha=0.5)
for disturbed in (utils.contaminate(predicted, 0.1) for i in range(5)):
    p = exaust(levmarq([BasinPolygon2DGz(x, z, disturbed, 0, npoints, density)] + regularization, initial))
    mpl.polygon(data.topolygon(p), '-r', linewidth=3, linealpha=0.5)
mpl.polygon(data.topolygon(estimate), '.-b', linewidth=2)
mpl.ylim(depths.max() + 1000, depths.min())
mpl.xlim(x.min(), x.max())
mpl.m2km()
mpl.subplot(2, 1, 1)
mpl.plot(x*0.001, gz, '.k')
mpl.plot(x*0.001, predicted, '-r')
```

## Example: Tomography

```python
class TravelTime(L2Norm):
    def __init__(self, ttime, srcs, recs, mesh):
        super(TravelTime, self).__init__(ttime)
        self._stored_jacobian = None
        self.srcs = srcs
        self.recs = recs
        self.mesh = mesh
        self.nparams = mesh.size
    
    def _precompute_jacobian(self):
        i, j, v = [], [], []
        for k, c in enumerate(mesh):
            column = seismic.ttime2d.straight([c], '', self.srcs, self.recs, velocity=1.)
            nonzero = numpy.flatnonzero(column)
            i.extend(nonzero)
            j.extend(k*numpy.ones_like(nonzero))
            v.extend(column[nonzero])
        return scipy.sparse.coo_matrix((v, (i, j)), (self.size, self.nparams)).tocsr()
        
    def jacobian(self, p):
        if self._stored_jacobian is None:
            self._stored_jacobian = self._precompute_jacobian()
        return self._stored_jacobian
    
    def predicted(self, p):
        if self._stored_jacobian is None:            
            self._stored_jacobian = self._precompute_jacobian()
        pred = safe_dot(self._stored_jacobian, p)
        if len(pred.shape) > 1:
            pred = numpy.array(pred.T).ravel()
        return pred        
```

```python
area = (0, 500000, 0, 500000)
shape = (100, 100)
model = mesher.SquareMesh(area, shape)
model.img2prop('logo.png', 4000, 10000, 'vp')
seed = 0 # Set the random seed so that points are the same everythime
src_loc = utils.random_points(area, 80, seed=seed)
rec_loc = utils.circular_points(area, 30, random=True, seed=seed)
srcs, recs = utils.connect_points(src_loc, rec_loc)
tts = seismic.ttime2d.straight(model, 'vp', srcs, recs, par=False)
tts, error = utils.contaminate(tts, 0.01, percent=True, return_stddev=True)
```

```python
mesh = mesher.SquareMesh(area, shape)
data = TravelTime(tts, srcs, recs, mesh)
print data.size, mesh.size
```
