# Sample implementation for the fatiando.inversion package refactor

This is an Object Oriented implementation of the inverse problem solvers.

```python
%pylab inline
from __future__ import division
import hashlib
import numpy
import matplotlib.pyplot as plt
import scipy.sparse
from numexpr import evaluate
from fatiando import gravmag, mesher, utils, seismic, gridder
from fatiando.vis import mpl
import time
```

```python
def solve(matrix, vector):
    if scipy.sparse.issparse(matrix) or scipy.sparse.issparse(vector):
        estimate, status = scipy.sparse.linalg.cgs(matrix, vector)
        if status >= 0:
            return estimate
        else:
            raise ValueError('CGS exited with input error')
    else:
        return numpy.linalg.solve(matrix, vector)
    
def dot(a, b):
    if scipy.sparse.issparse(a) or scipy.sparse.issparse(b):
        return a*b
    else:
        return numpy.dot(a, b)
    
def diagonal(matrix):
    if scipy.sparse.issparse(matrix):
        return numpy.array(matrix.diagonal())
    else:
        return numpy.diagonal(matrix).copy()
```

## Misfit functions

This implementation uses classes to define misfit functions.
These functions know how to optimize themselves using a variety of methods.
In order to do this, they must know how to calculate their gradient, Hessian, and value.
These things can be implemented in a generic way so that as little as possible is left for the user to implement.

```python
class Misfit(object):
    def __init__(self, nparams, islinear=False):
        self.init_report()
        self._cache = {}
        self.hasher = lambda x: hashlib.sha1(x).hexdigest()
        if islinear:
            self.fit = self.overdetermined
            self.islinear = True
        else:
            self.fit = self.levmarq
            self.islinear = False
        self.nparams = nparams
        
    def init_report(self):        
        self._report = {
            'method':'',
            'iterations':'N/A',
            'misfit/iteration':'N/A',
            'step search stagnation':'N/A',
            'step size/iteration':'N/A',
            'trial steps/iteration':'N/A',
            'preconditioned':'N/A'}
    
    def report(self, summary=True):
        text = '\n'.join([
            'method: %s' % (self._report['method']),
            'iterations: %s' % (str(self._report['iterations'])),
            'preconditioned: %s' % (str(self._report['preconditioned'])),
            'step search stagnation: %s' % (str(self._report['step search stagnation']))
            ])
        if not summary:
            text = '\n'.join([text,
                'misfit/iteration: %s' % (str(self._report['misfit/iteration'])),
                'step size/iteration: %s' % (str(self._report['step size/iteration'])),
                'trial steps/iteration: %s' % (str(self._report['trial steps/iteration']))
                ])
        return text
    
    def value(self, p):
        NotImplementedError("Misfit value not implemented") 
        
    def gradient(self, p):
        raise NotImplementedError("Gradient vector not implemented") 
        
    def hessian(self, p):
        raise NotImplementedError("Hessian matrix not implemented") 
        
    def overdetermined(self, precondition=True):
        self.init_report()
        self._report['method'] = 'Overdetermined Least Squares'
        self._report['preconditioned'] = precondition
        hessian = self.hessian(None)
        gradient = self.gradient(None)
        if precondition:
            diag = numpy.abs(diagonal(hessian))
            diag[diag < 10**-10] = 10**-10
            precond = scipy.sparse.diags(1./diag, 0).tocsr()
            del diag
            hessian = dot(precond, hessian)
            gradient = dot(precond, gradient)
        p = solve(hessian, -gradient)
        return p
    
    # Levemberg-Marquardt algorithm
    def levmarq(self, initial, maxit=30, maxsteps=10, lamb=10, dlamb=2, tol=10**-5, precondition=True):
        for p in self.ilevmarq(initial, maxit, maxsteps, lamb, dlamb, tol, precondition):
            continue
        return p
    
    def ilevmarq(self, initial, maxit=30, maxsteps=10, lamb=10, dlamb=2, tol=10**-5, precondition=True):        
        p = initial
        misfit = self.value(p)
        self.init_report()
        self._report['method'] = 'Levemberg-Marquardt'
        self._report['iterations'] = 0
        self._report['misfit/iteration'] = [misfit]
        self._report['step search stagnation'] = False
        self._report['step size/iteration'] = []
        self._report['trial steps/iteration'] = []
        self._report['preconditioned'] = precondition
        yield p
        for iteration in xrange(maxit):
            hessian = self.hessian(p)
            minus_gradient = -self.gradient(p)
            if precondition:
                diag = numpy.abs(diagonal(hessian))
                diag[diag < 10**-10] = 10**-10
                precond = scipy.sparse.diags(1./diag, 0).tocsr()
                del diag                
                hessian = dot(precond, hessian)
                minus_gradient = dot(precond, minus_gradient)
            stagnation = True
            diag = scipy.sparse.diags(diagonal(hessian), 0).tocsr()
            for step in xrange(maxsteps):
                newp = p + solve(hessian + lamb*diag, minus_gradient)
                newmisfit = self.value(newp)
                if newmisfit >= misfit:
                    if lamb < 10**15:
                        lamb *= dlamb
                else:
                    if lamb > 10**-15:
                        lamb /= dlamb
                    stagnation = False
                    break
            if stagnation:
                stop = True
                self._report['step search stagnation'] = True
            else:
                stop = newmisfit > misfit or abs((newmisfit - misfit)/float(misfit)) < tol
                p = newp
                misfit = newmisfit
                self._report['iterations'] += 1
                self._report['misfit/iteration'].append(newmisfit)
                self._report['step size/iteration'].append(lamb)
                self._report['trial steps/iteration'].append(step + 1)
                yield p
            if stop:
                break
                
    # Newton or Gauss-Newton method
    def newton(self, initial, maxit=30, tol=10**-5, precondition=True):
        for p in self.inewton(initial, maxit=maxit, tol=tol, precondition=precondition):
            continue
        return p
                
    def inewton(self, initial, maxit=30, tol=10**-5, precondition=True):        
        p = initial.astype(numpy.float)
        misfit = self.value(p)
        self.init_report()
        self._report['method'] = 'Newton'
        self._report['iterations'] = 0
        self._report['misfit/iteration'] = [misfit]
        self._report['preconditioned'] = precondition
        yield p
        for iteration in xrange(maxit):
            hessian = self.hessian(p)
            gradient = self.gradient(p)
            if precondition:
                diag = numpy.abs(diagonal(hessian))
                diag[diag < 10**-10] = 10**-10
                precond = scipy.sparse.diags(1./diag, 0).tocsr()
                del diag                
                hessian = dot(precond, hessian)
                gradient = dot(precond, gradient)
            p = p + solve(hessian, -gradient)
            newmisfit = self.value(p)
            if newmisfit > misfit or abs((newmisfit - misfit)/float(misfit)) < tol:
                break
            misfit = newmisfit
            self._report['iterations'] += 1
            self._report['misfit/iteration'].append(misfit)
            yield p
                        
    # Steepest descent with line search
    def steepest(self, initial, stepsize=0.1, maxsteps=30, maxit=500, tol=10**-5):
        for p in self.isteepest(initial, stepsize=stepsize, maxsteps=maxsteps, maxit=maxit, tol=tol):
            continue
        return p
            
    def isteepest(self, initial, stepsize=0.1, maxsteps=30, maxit=500, tol=10**-5):
        p = initial
        misfit = self.value(p)
        self.init_report()
        self._report['method'] = 'Steepest Descent'
        self._report['iterations'] = 0
        self._report['misfit/iteration'] = [misfit]
        self._report['step search stagnation'] = False
        self._report['step size/iteration'] = []
        self._report['trial steps/iteration'] = [] 
        yield p
        # This is a mistic parameter of the Armijo rule
        alpha = 10**(-4)
        for iteration in xrange(maxit):
            gradient = self.gradient(p) 
            # Calculate now to avoid computing inside the loop
            gradnorm = numpy.linalg.norm(gradient)**2
            stagnation = True
            # Determine the best step size
            for i in xrange(maxsteps):
                factor = stepsize**i
                newp = p - factor*gradient
                newmisfit = self.value(newp)
                if newmisfit - misfit < alpha*factor*gradnorm:
                    stagnation = False
                    break
            if stagnation:
                stop = True
                self._report['step search stagnation'] = True
            else:
                stop = (abs((newmisfit - misfit)/float(misfit)) < tol)
                p = newp
                misfit = newmisfit
                self._report['iterations'] += 1
                self._report['misfit/iteration'].append(misfit)
                self._report['step size/iteration'].append(factor)
                self._report['trial steps/iteration'].append(i + 1)
                yield p
            if stop:
                break
                
    # Ant Colony Optimization for Continuous Domains
    def acor(self, bounds, nants=None, archive_size=None, maxit=1000, diverse=0.5, evap=0.85, 
             cache=False, reset=False, seed=None):
        optimizer = self.iacor(bounds, nants, archive_size, maxit=maxit, diverse=diverse, 
                               evap=evap, cache=cache, reset=reset, seed=seed)
        for p in optimizer:
            continue
        return p
                
    def iacor(self, bounds, nants=None, archive_size=None, maxit=1000, diverse=0.5, evap=0.85, 
              cache=False, reset=False, seed=None):
        numpy.random.seed(seed)
        if nants is None:
            nants = self.nparams
        if archive_size is None:
            archive_size = 10*nants
        bounds = numpy.array(bounds)
        if bounds.size == 2:
            low, high = bounds
            archive = numpy.random.uniform(low, high, (archive_size, self.nparams))
        else:
            archive = numpy.empty((archive_size, self.nparams))
            bounds = bounds.reshape((self.nparams, 2))
            for i, bound in enumerate(bounds):
                low, high = bound
                archive[:, i] = numpy.random.uniform(low, high, archive_size)
        trail = numpy.fromiter((self.value(p) for p in archive), dtype=numpy.float)
        order = numpy.argsort(trail)
        archive = [archive[i] for i in order]
        trail = trail[order].tolist()
        yield archive[0]
        aux =1/(diverse*archive_size*numpy.sqrt(2*numpy.pi))
        weights = aux*numpy.exp(-numpy.arange(archive_size)**2/(2*diverse**2*archive_size**2))
        weights /= numpy.sum(weights)
        for iteration in xrange(maxit):
            for k in xrange(nants):
                # Sample the propabilities to produce new estimates
                ant = numpy.empty(self.nparams)
                # 1. Choose a pdf from the archive
                pdf = numpy.searchsorted(
                            numpy.cumsum(weights),
                            numpy.random.uniform())
                for i in xrange(self.nparams):
                    # 2. Get the parameters of the chosen pdf
                    mean = archive[pdf][i]                    
                    std = evap*numpy.sum(abs(p[i] - archive[pdf][i]) for p in archive)/(archive_size - 1)
                    # 3. Sample the pdf until the samples are in bounds
                    for atempt in xrange(100):
                        ant[i] = numpy.random.normal(mean, std)
                        if bounds.size == 2:
                            low, high = bounds
                        else:
                            low, high = bounds[i]
                        if ant[i] >= low and ant[i] <= high:
                            break
                pheromone = self.value(ant)                
                # Place the new estimate in the archive
                place = numpy.searchsorted(trail, pheromone)
                if place == archive_size:
                    continue
                trail.insert(place, pheromone)
                trail.pop()
                archive.insert(place, ant)
                archive.pop()
            yield archive[0]                    
                
    # Overload some operators
    def __add__(self, other):
        if not isinstance(other, Misfit):
            raise TypeError('Can only add derivatives of the Misfit class')
        goal = GoalFunction()
        if isinstance(other, GoalFunction):
            goal.merge(other)
        else:
            goal.add_misfit(other)
        if isinstance(self, GoalFunction):
            goal.merge(self)
        else:
            goal.add_misfit(self)
        return goal
    
    def __mul__(self, other):
        if not isinstance(other, int) and not isinstance(other, float):
            raise TypeError('Can only multiply a Misfit by a float or int')
        return GoalFunction([(other, self)])
    
    def __rmul__(self, other):
        return self.__mul__(other)
                
# A Goal function is the weighted sum of many misfits
class GoalFunction(Misfit):
    def __init__(self, misfits=None):
        super(GoalFunction, self).__init__(nparams=None, islinear=False)
        self.misfits = []
        if misfits is not None:
            for mu, misfit in misfits:
                self.add_misfit(misfit, regul_param=mu)            
            
    def add_misfit(self, misfit, regul_param=1):
        nparams = misfit.nparams
        if self.nparams is not None:
            if numpy.any([nparams != m.nparams for _, m in self.misfits]):
                raise ValueError(
                    'Misfit function must have %d parameters, not %d'
                    % (self.nparams, nparams))
        else:
            self.nparams = nparams
        self.misfits.append([regul_param, misfit])
        if numpy.any([not m.islinear for _, m in self.misfits]):
            self.islinear = False
            self.fit = self.levmarq
        else:
            self.islinear = True
            self.fit = self.overdetermined
        return self
        
    def merge(self, goalfunc):
        for mu, misfit in goalfunc:
            self.add_misfit(misfit, regul_param=mu)
        return self
    
    # Can increment instead of add_misfit or merge
    def __iadd__(self, other):
        if not isinstance(other, Misfit):
            raise TypeError('Can only add derivatives of the Misfit class')
        if isinstance(other, GoalFunction):
            self.merge(other)
        else:
            self.add_misfit(other)
        return self
    
    # Make it behave like a list, sort of
    def __len__(self):
        return len(self.misfits)

    def __iter__(self):
        self.index = 0
        return self

    def __getitem__(self, index):
        return self.misfits[index]

    def next(self):
        if self.index >= len(self.misfits):
            raise StopIteration
        mu, misfit = self.__getitem__(self.index)
        self.index += 1
        return mu, misfit
            
    def __str__(self):
        description = '\n'.join(
            ['Goal function composed of:'] +
            ['  %g*%s' % (mu, str(misfit.__class__))
             for mu, misfit in self.misfits])
        return description
        
    def value(self, p):
        return sum(mu*misfit.value(p) for mu, misfit in self.misfits)
        
    def gradient(self, p):
        return sum(mu*misfit.gradient(p) for mu, misfit in self.misfits)
        
    def hessian(self, p):
        return sum(mu*misfit.hessian(p) for mu, misfit in self.misfits)
    
    def predicted(self, p):
        pred = []
        for mu, misfit in self.misfits:
            if callable(getattr(misfit, 'predicted', None)):
                pred.append(misfit.predicted(p))
        if len(pred) == 1:
            pred = pred[0]
        return pred
```

The $\ell_e$-norm data misfit function (for least-squares) is a kind of misfit function.

```python
class L2Norm(Misfit):
    def __init__(self, data, nparams, weights=None, islinear=False):
        super(L2Norm, self).__init__(nparams, islinear=islinear)
        self.data = data
        self.ndata = len(data)
        self._cache['predicted'] = {'hash':'', 'array':None}
        self._cache['jacobian'] = {'hash':'', 'array':None}
        self._cache['hessian'] = {'hash':'', 'array':None}
        if weights is not None:
            self.set_weights(weights)
        else:
            self.weights = None
                
    def set_data(self, data):
        self.data = data
        return self
    
    def set_weights(self, weights):
        self.weights = scipy.sparse.diags(weights, 0)
        # Weights change the Hessian
        self._cache['hessian'] = {'hash':'', 'array':None}
        return self
            
    def residuals(self, p):
        return self.data - self.predicted(p)
            
    def predicted(self, p):
        if p is None:
            pred = 0
        else:
            hash = self.hasher(p)
            if hash != self._cache['predicted']['hash']:
                self._cache['predicted']['array'] = self._get_predicted(p)
                self._cache['predicted']['hash'] = hash            
            pred = self._cache['predicted']['array']
        return pred
    
    def jacobian(self, p):
        if self.islinear:
            hash = ''
        else:
            hash = self.hasher(p)
        if hash != self._cache['jacobian']['hash'] or self._cache['jacobian']['array'] is None:
            self._cache['jacobian']['array'] = self._get_jacobian(p)
            self._cache['jacobian']['hash'] = hash
        return self._cache['jacobian']['array']
                
    def value(self, p):        
        if self.weights is None:    
            return numpy.linalg.norm(self.data - self.predicted(p))**2/self.ndata
        else:
            return numpy.sum(self.weights*((self.data - self.predicted(p))**2))/self.ndata 
    
    def hessian(self, p):
        if self.islinear and self._cache['hessian']['array'] is not None:
            hessian = self._cache['hessian']['array']
        else:
            jacobian = self.jacobian(p)
            if self.weights is None:
                hessian = (2/self.ndata)*dot(jacobian.T, jacobian)
            else:
                hessian = (2/self.ndata)*dot(jacobian.T, self.weights*jacobian)
            if self.islinear:                
                self._cache['hessian']['array'] = hessian
        return hessian
    
    def gradient(self, p):
        jacobian = self.jacobian(p)
        if self.weights is None:
            grad = (-2/self.ndata)*dot(jacobian.T, self.data - self.predicted(p))
        else:
            grad = (-2/self.ndata)*dot(jacobian.T, self.weights*(self.data - self.predicted(p)))
        # Check if the gradient isn't a one column matrix
        if len(grad.shape) > 1:
            # Need to convert it to a 1d array so that hell won't break loose
            grad = numpy.array(grad).ravel()
        return grad
        
    def _get_predicted(self, p):
        raise NotImplementedError("Predicted data not implemented") 
        
    def _get_jacobian(self, p):
        raise NotImplementedError("Jacobian matrix not implemented")
```

Regularizing functions are also misfit functions:

```python
class Damping(Misfit):
    def __init__(self, nparams):
        super(Damping, self).__init__(nparams, islinear=True)
        
    def hessian(self, p):
        # This is cheap so there is no need to cache it
        return 2*scipy.sparse.identity(self.nparams).tocsr()
    
    def gradient(self, p):
        if p is None:
            grad = 0
        else:
            grad = 2*p
        return grad
        
    def value(self, p):
        return numpy.linalg.norm(p)**2
    
class Smoothness(Misfit):
    def __init__(self, fdmat):
        super(Smoothness, self).__init__(fdmat.shape[1], islinear=True)
        self._cache['hessian'] = {'hash':'', 'array':2*dot(fdmat.T, fdmat)}
        
    def hessian(self, p):
        return self._cache['hessian']['array']
    
    def gradient(self, p):
        if p is None:
            grad = 0
        else:
            grad = dot(self.hessian(p), p)
        return grad
               
    def value(self, p):
        # Need to divide by 2 because the hessian is 2*R.T*R
        return numpy.sum(p*dot(self.hessian(p), p))/2.
    
class Smoothness1D(Smoothness):
    def __init__(self, npoints):
        super(Smoothness1D, self).__init__(fd1d(npoints))
        
class Smoothness2D(Smoothness):
    def __init__(self, shape):
        super(Smoothness2D, self).__init__(fd2d(shape))
                
class TotalVariation(Misfit):
    def __init__(self, beta, fdmat):
        super(TotalVariation, self).__init__(fdmat.shape[1], islinear=False)
        self.beta = beta
        self._fdmat = fdmat
        
    def value(self, p):
        return numpy.linalg.norm(dot(self._fdmat, p), 1)
    
    def hessian(self, p):
        derivs = dot(self._fdmat, p)
        q_matrix = scipy.sparse.diags(self.beta/((derivs**2 + self.beta)**1.5), 0).tocsr()
        return dot(self._fdmat.T, q_matrix*self._fdmat)
    
    def gradient(self, p):
        derivs = dot(self._fdmat, p)
        q = derivs/numpy.sqrt(derivs**2 + self.beta)
        grad = dot(self._fdmat.T, q)
        if len(grad.shape) > 1:
            grad = numpy.array(grad.T).ravel()
        return grad
    
class TotalVariation1D(TotalVariation):
    def __init__(self, beta, npoints):
        super(TotalVariation1D, self).__init__(beta, fd1d(npoints))
    
class TotalVariation2D(TotalVariation):
    def __init__(self, beta, shape):
        super(TotalVariation2D, self).__init__(beta, fd2d(shape))
            
def fd1d(size):
    i = range(size - 1) + range(size - 1)
    j = range(size - 1) + range(1, size)
    v = [1]*(size - 1) + [-1]*(size - 1)
    return scipy.sparse.coo_matrix((v, (i, j)), (size - 1, size)).tocsr()

def fd2d(shape):
    ny, nx = shape
    nderivs = (nx - 1)*ny + (ny - 1)*nx
    I, J, V = [], [], []
    deriv = 0
    param = 0
    for i in xrange(ny):
        for j in xrange(nx - 1):
            I.extend([deriv, deriv])
            J.extend([param, param + 1])
            V.extend([1, -1])
            deriv += 1
            param += 1
        param += 1
    param = 0
    for i in xrange(ny - 1):
        for j in xrange(nx):
            I.extend([deriv, deriv])
            J.extend([param, param + nx])
            V.extend([1, -1])
            deriv += 1
            param += 1
    return scipy.sparse.coo_matrix((V, (I, J)), (nderivs, nx*ny)).tocsr()
```

## Example use: polynomial regression

This is a problem specific misfit function that wraps the data and auxiliary values. It also knows how to compute the Jacobian matrix and predicted data for the case of a second order regression.

```python
class PolyData(L2Norm):
    def __init__(self, x, y, weights=None):
        super(PolyData, self).__init__(y, nparams=3, weights=weights, islinear=True)
        self.x = x
        
    def _get_jacobian(self, p):
        return numpy.transpose([self.x**2, self.x, numpy.ones_like(self.x)])

    def _get_predicted(self, p):
        return dot(self.jacobian(p), p)
```

Testing this on some synthetic data:

```python
x = numpy.linspace(0, 10, 100)
numpy.random.seed(1)
a, b, c = 10, -100, 30000
y = a*x**2 + b*x + c + numpy.random.normal(scale=20, size=len(x))
data = PolyData(x, y)
coefs = data.fit()
plt.figure()
# Use the linear solver
print "LINEAR:"
print 'estimated:', coefs.tolist()
print 'true:', [a, b, c]
print data.report()
plt.plot(x, y, '.k', x, data.predicted(coefs), '-r', linewidth=3)
# Run through the non-linear optimizer
print "\nNON-LINEAR: Levemberg-Marquardt"
coefs = data.levmarq(numpy.array([0, 0, 0]))
print 'estimated:', coefs.tolist()
print 'true:', [a, b, c]
print data.report()
plt.plot(x, data.predicted(coefs), '--b', linewidth=3)
print "\nNON-LINEAR: ACO-R"
coefs = data.acor([0, 50, -500, 0, 10000, 100000])
print 'estimated:', coefs.tolist()
print 'true:', [a, b, c]
print data.report()
plt.plot(x, data.predicted(coefs), '--g', linewidth=3)
```

## Example: Gaussian fit

Fit an equation of the form:

$$ f(x) = a\exp(-b(x + c)^{2}) + d\exp(-e(x + f)^{2}) $$

```python
class ExpData(L2Norm):
    def __init__(self, x, y):
        super(ExpData, self).__init__(y, nparams=6, islinear=False)
        self.x = x
    
    def _get_jacobian(self, p):
        jac = numpy.empty((self.ndata, self.nparams))
        jac[:,0], jac[:,1], jac[:,2] = self._derivs(p[:3])
        jac[:,3], jac[:,4], jac[:,5] = self._derivs(p[3:])
        return jac
    
    def _derivs(self, coefs):
        a, b, c = coefs
        derivs = [
            numpy.exp(-b*(self.x + c)**2),
            # Do a numerical derivative for these two because I'm too lazy to do it by hand
            (a*numpy.exp(-(b + 0.005)*(self.x + c)**2) - a*numpy.exp(-(b - 0.005)*(self.x + c)**2))/0.01,
            (a*numpy.exp(-(b)*(self.x + c + 0.05)**2) - a*numpy.exp(-(b)*(self.x + c - 0.05)**2))/0.1]
        return derivs
    
    def _get_predicted(self, p):
        a, b, c, d, e, f = p
        return a*numpy.exp(-b*(self.x + c)**2) + d*numpy.exp(-e*(self.x + f)**2)
            
```

```python
x = numpy.linspace(0, 10, 1000)
numpy.random.seed(1)
a, b, c = 100, 0.1, -2
d, e, f = 30, 2, -7
y = a*numpy.exp(-b*(x + c)**2) + d*numpy.exp(-e*(x + f)**2) + numpy.random.normal(scale=5, size=len(x))
data = ExpData(x, y)
initial = numpy.array([1, 1, 1, 1, 1, -9])
coefs = data.fit(initial)
print 'estimate:', coefs.tolist()
print 'true:', [a, b, c, d, e, f]
print data.report()
plt.plot(x, y, '.k', x, data.predicted(coefs), '-r', linewidth=4)
coefs = data.acor([0, 200, 0.0001, 10, -10, 0]*2, evap=0.5)
print 'estimate:', coefs.tolist()
print 'true:', [a, b, c, d, e, f]
#print data.report()
plt.plot(x, data.predicted(coefs), '--b', linewidth=4)
```

## Example: 2D gravity inversion for the relief of a basin

```python
class BasinPolygon2DGz(L2Norm):
    def __init__(self, x, z, gz, height, npoints, density):
        super(BasinPolygon2DGz, self).__init__(gz, nparams=npoints, islinear=False)
        self.x, self.z = x, z
        self.height = height
        self.props = {'density':density}
        self._left, self._right = [[x.min(), -height]], [[x.max(), -height]]
        self.set_npoints(npoints)
    
    def set_npoints(self, npoints):
        self._xs = numpy.linspace(x.min(), x.max(), npoints + 2)[::-1][1:-1]
        self.nparams = npoints
        return self
        
    def toverts(self, p):
        return self._right + numpy.transpose([self._xs, p]).tolist() + self._left
    
    def topolygon(self, p):
        return mesher.Polygon(self.toverts(p), self.props)
    
    def _get_jacobian(self, p):
        verts = self.toverts(p)
        x, z = self.x, self.z
        xs = self._xs
        dz = 10.
        jac = numpy.zeros((self.ndata, self.nparams))
        for i in xrange(self.nparams):
            difference = mesher.Polygon([verts[i+2],[xs[i],p[i]-dz],verts[i],[xs[i],p[i]+dz]], self.props)
            jac[:,i] = numpy.nan_to_num(gravmag.talwani.gz(x, z, [difference]))/(2.*dz)
        return jac
        
    def _get_predicted(self, p):
        return gravmag.talwani.gz(self.x, self.z, [self.topolygon(p)])
```

```python
x = numpy.linspace(0, 100000, 100)
z = -100*numpy.ones_like(x)
xs = numpy.linspace(0, 100000, 200)[::-1]
depths = -10**-15*(xs - 50000)**4 + 6000 - 5000*numpy.exp(-(xs - 70000)**2/(10000**2))
depths -= depths.min()
depths[(xs > 30000) & (xs < 40000)] -= 3000
density = -500
model = mesher.Polygon(numpy.transpose([xs, depths]), {'density':density})
gz = utils.contaminate(gravmag.talwani.gz(x, z, [model]), 0.1, seed=20)
npoints = 100
data = BasinPolygon2DGz(x, z, gz, 0, npoints, density)

mpl.figure(figsize=(10,6))
mpl.subplot(2, 1, 2)
mpl.polygon(model, fill='gray', alpha=0.5)
mpl.ylim(depths.max() + 1000, depths.min())
mpl.xlim(x.min(), x.max())
mpl.m2km()
mpl.subplot(2, 1, 1)
mpl.plot(x*0.001, gz, '.k')
```

Make the misfit class.

## Example: Tomography

```python
class TravelTime(L2Norm):
    def __init__(self, ttime, srcs, recs, mesh):
        super(TravelTime, self).__init__(ttime, nparams=mesh.size, islinear=True)
        self.srcs = srcs
        self.recs = recs
        self.mesh = mesh
    
    def _get_jacobian(self, p):
        i, j, v = [], [], []
        for k, c in enumerate(self.mesh):
            column = seismic.ttime2d.straight([c], '', self.srcs, self.recs, velocity=1.)
            nonzero = numpy.flatnonzero(column)
            i.extend(nonzero)
            j.extend(k*numpy.ones_like(nonzero))
            v.extend(column[nonzero])
        return scipy.sparse.coo_matrix((v, (i, j)), (self.ndata, self.nparams)).tocsr()
            
    def _get_predicted(self, p):
        pred = dot(self.jacobian(p), p)
        if len(pred.shape) > 1:
            pred = numpy.array(pred.T).ravel()
        return pred        
```

```python
area = (0, 500000, 0, 500000)
shape = (100, 100)
model = mesher.SquareMesh(area, shape)
model.img2prop('logo.png', 4000, 10000, 'vp')
seed = 0 # Set the random seed so that points are the same everythime
src_loc = utils.random_points(area, 80, seed=seed)
rec_loc = utils.circular_points(area, 30, random=True, seed=seed)
srcs, recs = utils.connect_points(src_loc, rec_loc)
tts = seismic.ttime2d.straight(model, 'vp', srcs, recs, par=False)
tts, error = utils.contaminate(tts, 0.01, percent=True, return_stddev=True)
```

```python
mesh = mesher.SquareMesh(area, shape)
data = TravelTime(tts, srcs, recs, mesh)
print data.ndata, mesh.size
```

Use smoothness:

```python
smooth = GoalFunction()
smooth.add_misfit(data).add_misfit(Smoothness2D(mesh.shape), regul_param=1000000)
estimate = smooth.fit()
residuals = data.residuals(estimate)
mesh.addprop('vp', seismic.srtomo.slowness2vel(estimate))
print "Assumed error: %g" % (error)
print "Standard deviation of residuals: %g" % (numpy.std(residuals))
print smooth.report()

mpl.figure(figsize=(18, 4))
mpl.subplot(1, 3, 1)
mpl.axis('scaled')
mpl.title('Vp synthetic model of the Earth')
mpl.squaremesh(model, prop='vp', cmap=mpl.cm.seismic)
cb = mpl.colorbar()
cb.set_label('Velocity')
mpl.points(src_loc, '*y', label="Sources")
mpl.points(rec_loc, '^r', label="Receivers")
mpl.legend(loc='lower left', shadow=True, numpoints=1, prop={'size':10})
mpl.m2km()
mpl.subplot(1, 3, 2)
mpl.axis('scaled')
mpl.title('Tomography result')
mpl.squaremesh(mesh, prop='vp', vmin=4000, vmax=10000, cmap=mpl.cm.seismic)
cb = mpl.colorbar()
cb.set_label('Velocity')
mpl.m2km()
mpl.subplot(1, 3, 3)
mpl.grid()
mpl.title('Residuals (data with %.4f s error)' % (error))
mpl.hist(residuals, color='gray', bins=10)
mpl.xlabel("seconds")
```

Use total variation:

```python
sharp = 1000*TotalVariation2D(10**-7, mesh.shape)
sharp += data
initial = 0.0005*numpy.ones(mesh.size)
# TotalVariation is non-linear, so fit is levmarq
estimate = sharp.fit(initial)
residuals = data.residuals(estimate)
mesh.addprop('vp', seismic.srtomo.slowness2vel(estimate))
print "Assumed error: %g" % (error)
print "Standard deviation of residuals: %g" % (numpy.std(residuals))
print sharp.report()

mpl.figure(figsize=(18, 4))
mpl.subplot(1, 3, 1)
mpl.axis('scaled')
mpl.title('Vp synthetic model of the Earth')
mpl.squaremesh(model, prop='vp', cmap=mpl.cm.seismic)
cb = mpl.colorbar()
cb.set_label('Velocity')
mpl.points(src_loc, '*y', label="Sources")
mpl.points(rec_loc, '^r', label="Receivers")
mpl.legend(loc='lower left', shadow=True, numpoints=1, prop={'size':10})
mpl.m2km()
mpl.subplot(1, 3, 2)
mpl.axis('scaled')
mpl.title('Tomography result')
mpl.squaremesh(mesh, prop='vp', vmin=4000, vmax=10000, cmap=mpl.cm.seismic)
cb = mpl.colorbar()
cb.set_label('Velocity')
mpl.m2km()
mpl.subplot(1, 3, 3)
mpl.grid()
mpl.title('Residuals (data with %.4f s error)' % (error))
mpl.hist(residuals, color='gray', bins=10)
mpl.xlabel("seconds")
```

## Example: 2D epicenter determination

```python
class EpicTravelTime(L2Norm):
    def __init__(self, traveltimes, recs, vp, vs):
        super(EpicTravelTime, self).__init__(traveltimes, nparams=2, islinear=False)
        self.x, self.y = numpy.transpose(recs)
        self.alpha = 1./vs - 1./vp
        
    def _get_predicted(self, p):
        x, y = p
        return self.alpha*numpy.sqrt((self.x - x)**2 + (self.y - y)**2)

    def _get_jacobian(self, p):
        x, y = p
        sqrt = numpy.sqrt((self.x - x)**2 + (self.y - y)**2)
        jacobian = numpy.transpose([
                -self.alpha*(self.x - x)/sqrt,
                -self.alpha*(self.y - y)/sqrt])
        return jacobian

def map2dgoal(area, shape, goal):
    n = shape[0]*shape[1]
    paramspace = numpy.empty((n, 2))
    paramspace[:,0], paramspace[:,1] = gridder.regular(area, shape)
    goals = numpy.fromiter((data.value(p) for p in paramspace), dtype=numpy.float)
    return paramspace[:,0], paramspace[:,1], goals
```
