# A layered cube model of the Earth

This prototype defines a class that describes a model of the Earth as a layered cube. This is a very rough approximation but it  serves many teaching purposes by allowing us the compute theorical gravity anomalies, isostasy, etc.

```python
%matplotlib inline
from __future__ import division
from fatiando.gravmag import prism
from fatiando.mesher import Prism
from fatiando.vis import mpl, myv
from fatiando import utils
import fatiando
import numpy as np
from IPython.display import Image
```

```python
fatiando.__version__
```

The `CubeEarth` class defines the layered cube model. The default values for the thickness and density of each layer are based on the [PREM](http://geophysics.ou.edu/solid_earth/prem.html). The class defines many properties that calculate useful things, like Moho depth, density contrast profile, etc. It also behaves as an iterator, so it can be passed to `myv.prisms` and `gravmag.prism.gz` like a `PrismMesh`.

```python
class CubeEarth(object):
    
    _normal_densities = np.array([2670, 3500, 5000, 10000, 13000])
    _normal_thickness = np.array([25e3, 500e3, 2200e3, 2200e3, 1200e3])
    
    def __init__(self, density=None, thickness=None):            
        self._densities = density
        self._thickness = thickness
        assert len(self.density) == len(self.thickness), \
            'Density and thickness profiles have different lengths: {} and {}'.format(
                len(self.density), len(self.thickness))
        self._i = 0
        self.prisms = [Prism(-r, r, -r, r, -r, r, {'density': c})
                       for r, c in zip(self.radius, self.density_contrast)]
        
    def __add__(self, obj):
        return self.prisms + obj
    
    def __radd__(self, obj):
        return self.prisms + obj
            
    def __len__(self):
        return self.size
    
    def __iter__(self):
        self._i = 0
        return self
    
    def next(self):
        if self._i >= self.size:
            raise StopIteration
        p = self[self._i]
        self._i += 1
        return p
        
    def __getitem__(self, item):
        return self.prisms[item]
        
    @property
    def size(self):
        return len(self.thickness)
    
    @property
    def density(self):
        if self._densities is None:
            return self._normal_densities
        else:
            return self._densities
        
    @property
    def thickness(self):
        if self._thickness is None:
            return self._normal_thickness
        else:
            return self._thickness

    @property
    def radius(self):
        r = np.fromiter((sum(self.thickness[i:]) for i in xrange(self.size)), 
                        dtype=np.float)
        return r
    
    @property
    def earth_radius(self):
        return sum(self.thickness)
    
    @property
    def bounds(self):
        return [-self.earth_radius, self.earth_radius]*3
    
    @property
    def density_contrast(self):
        dc = np.empty(self.size, dtype=np.float)
        dc[0] = self.density[0]
        dc[1:] = self.density[1:] - self.density[0:-1]
        return dc
        
    @property
    def moho(self):
        return self.radius[1]
    
    @property
    def moho_depth(self):
        return self.thickness[0]
```

## Example

Calling `CubeEarth` without any arguments creates a "normal Earth" model (our reference model, like the ellipsoid).

```python
cb = CubeEarth()
```

By default, the `'density'` property of `CubeEarth` components is actually the density constrast. This makes it easier to pass it to `fatiando.gravmag.prism` functions and get the correct values.

```python
scene = myv.figure(size=(400, 400), zdown=False)
plot = myv.prisms(cb, 'density', opacity=0.5, linewidth=1)
plot.module_manager.scalar_lut_manager.lut_mode = 'Reds'
myv.show()
```

```python
Image(filename='cube-normal-earth.png')
```

Now that we have our "normal Earth", we can make a "real Earth" by adding a topographic high and a corresponding crustal thickening. We'll assume that the total  mass remained unchanged because the added mass is very small.

```python
topo_height = 5e3
xmin, xmax = -100e3, 100e3
topo = Prism(xmin, xmax, xmin, xmax, cb.earth_radius, cb.earth_radius + topo_height, 
             {'density': cb.density[0]})
thickening = Prism(xmin, xmax, xmin, xmax, cb.moho - 25e3, cb.moho,
                   {'density': cb.density[0] - cb.density[1]})
real_earth = cb + [topo, thickening]
```

```python
scene = myv.figure(size=(400, 400), zdown=False)
plot = myv.prisms(real_earth, 'density', opacity=0.5, linewidth=1)
plot.module_manager.scalar_lut_manager.lut_mode = 'Reds'
myv.show()
```

Now  we can compute the real and normal gravity on top of topography.

```python
x = np.linspace(-200e3, 200e3, 500)
y = np.zeros_like(x)
z = cb.earth_radius*np.ones_like(x)
z[(x >= xmin) & (x <= xmax)] = cb.earth_radius + topo_height
```

```python
def gravity(x, y, z, model):
    gx = prism.gx(x, y, z, model)
    gy = prism.gy(x, y, z, model)
    gz = prism.gz(x, y, z, model)
    grav = np.sqrt(gx**2 + gy**2 + gz**2)
    return grav
```

```python
real_g = gravity(x, y, z, real_earth)
gamma = gravity(x, y, z, cb)
```

```python
mpl.figure(figsize=(14, 3))
mpl.plot(x, real_g, '-k', label='Real')
mpl.plot(x, gamma, '-r', label='Normal')
mpl.legend(loc='lower right')
mpl.xlabel('x (m)')
mpl.ylabel('Gravity (mGal)')
```

The free-air anomaly (or gravity disturbance) is the real gravity minus the normal gravity computed at the same point.

```python
fa = real_g - gamma
```

```python
mpl.figure(figsize=(14, 3))
mpl.plot(x, fa, '-k')
mpl.xlabel('x (m)')
mpl.ylabel('Free-air anomaly (mGal)')
```
